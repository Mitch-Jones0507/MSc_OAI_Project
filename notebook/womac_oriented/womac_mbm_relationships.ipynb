# WOMAC MBM Relationships

### Objective:
To explore the predictive capacity that MOAKS bone marrow lesion (MBM) variables at baseline, 12-months, and a change between baseline and 12-months, have on the two-dimensional vector containing WOMAC functional disability and stiffness (WOMADL, WOMSTF) scores - as well as their principal components - at a 12-month time period, 24-month time period, and a change between 12 and 24-months.

### Results:
Almost no relationships between MBM and WOMAC.
