# MBM Variable Relationships (WOMAC Dataset)

From the merged dataframe, moaks_shared_womac, 3077 unique knees are scored on MOAKS and WOMAC scales. Here, all MBM (MOAKS BML) scores are analysed through PCA for principal components that may explain their underlying relationships.

## Baseline MBM Variable PCA on WOMAC Drop Dataset

```python
from IPython.display import display
import pandas as pd
from scripts.mbm.womac_oriented.mbm_var_relationships_womac import v00_moaks_shared_womac_drop_mbm_pc_variance, v00_moaks_shared_womac_drop_mbm_pc_variance_ratio, v00_moaks_shared_womac_drop_mbm_pc_loadings, v00_moaks_shared_womac_mbm_columns

pc_df = pd.DataFrame({
    'PC': [f'PC{i+1}' for i in range(len(v00_moaks_shared_womac_drop_mbm_pc_variance))],
    'Variance': v00_moaks_shared_womac_drop_mbm_pc_variance,
    'Proportion of Total': v00_moaks_shared_womac_drop_mbm_pc_variance_ratio
})

loadings_df = pd.DataFrame(v00_moaks_shared_womac_drop_mbm_pc_loadings,index=v00_moaks_shared_womac_mbm_columns,columns=[f'PC{i+1}' for i in range(len(v00_moaks_shared_womac_drop_mbm_pc_variance))])

display(pc_df)
display(loadings_df)
```
