# K-L Distribution

```python
import pandas as pd
import plotly.express as px
from IPython.display import display
from missing_imputation import moaks_shared_kl_drop_df

moaks_shared_kl_drop_df = moaks_shared_kl_drop_df.copy()
moaks_shared_kl_drop_df['ΔXRKL'] = moaks_shared_kl_drop_df['V03XRKL'] - moaks_shared_kl_drop_df['V01XRKL']

moaks_shared_kl_drop_frequency = pd.DataFrame({
    variable: moaks_shared_kl_drop_df[variable].value_counts().reindex(range(0,5),fill_value=0)
    for variable in ['V01XRKL','V03XRKL','ΔXRKL']
})

print('Frequency of K-L scores (0-4) in V01XRKL, V03XRKL and frequency of change of K-L score')
display(moaks_shared_kl_drop_frequency)

print('Percentage distribubtion of K-L scores (0-4) in V01XRKL, V03XRKL and frequency of change of K-L score')
moaks_shared_kl_drop_percentage = moaks_shared_kl_drop_frequency / moaks_shared_kl_drop_frequency.sum() * 100
display(moaks_shared_kl_drop_percentage.round(2))

v01_moaks_shared_kl_drop_histogram = px.histogram(moaks_shared_kl_drop_df, x='V01XRKL', nbins=20)
v01_moaks_shared_kl_drop_histogram.show()

v03_moaks_shared_kl_drop_histogram = px.histogram(moaks_shared_kl_drop_df, x='V03XRKL', nbins=20)
v03_moaks_shared_kl_drop_histogram.show()

moaks_shared_kl_drop_change_histogram = px.histogram(moaks_shared_kl_drop_df,x='ΔXRKL',nbins=20)
moaks_shared_kl_drop_change_histogram.show()
```

```python
from missing_imputation import moaks_shared_kl_median_df

moaks_shared_kl_median_df = moaks_shared_kl_median_df.copy()
moaks_shared_kl_median_df['ΔXRKL'] = moaks_shared_kl_median_df['V03XRKL'] - moaks_shared_kl_median_df['V01XRKL']

moaks_shared_kl_median_frequency = pd.DataFrame({
    variable: moaks_shared_kl_median_df[variable].value_counts().reindex(range(0,5),fill_value=0)
    for variable in ['V01XRKL','V03XRKL','ΔXRKL']
})

print('Frequency of K-L scores (0-4) in V01XRKL, V03XRKL and frequency of change of K-L score')
display(moaks_shared_kl_median_frequency)

print('Percentage distribubtion of K-L scores (0-4) in V01XRKL, V03XRKL and frequency of change of K-L score')
moaks_shared_kl_median_percentage = moaks_shared_kl_median_frequency / moaks_shared_kl_median_frequency.sum() * 100
display(moaks_shared_kl_median_percentage.round(2))

v01_moaks_shared_kl_median_histogram = px.histogram(moaks_shared_kl_median_df, x='V01XRKL',nbins=20)
v01_moaks_shared_kl_median_histogram.show()

v03_moaks_shared_kl_median_histogram = px.histogram(moaks_shared_kl_median_df, x='V03XRKL',nbins=20)
v03_moaks_shared_kl_median_histogram.show()

moaks_shared_kl_median_change_histogram = px.histogram(moaks_shared_kl_median_df,x='ΔXRKL',nbins=20)
moaks_shared_kl_median_change_histogram.show()
```
