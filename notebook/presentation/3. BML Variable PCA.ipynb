# Step 3: BML Variable PCA

```python
from IPython.display import display, Markdown
```

### Objective
- Explore latent structure of the BML variable matrix
- Quantify the extent to which variance is distributed across BML variables

### Statistical Tests
- Principal Component Analysis

### Inputs and Outputs
- 45 BML variables at baseline from full dataset, ungrouped (Input)

- Principal components of BML variables (Output)

## 3.1 BML Variable PCA at Baseline
At baseline there are 45 BML variables spanning across 15 subregions and 3 tests (size, number, percentage). The baseline MOAKS dataset contains 2879 unique knees for this analysis.

```python
from scripts.mbm.kl_oriented.mbm_var_relationships_kl import v00_moaks_shared_kl_drop_mbm_pc_df, v00_moaks_shared_kl_drop_mbm_loadings_df

display(Markdown('### 3.1.1 BML Variable Principal Components'))
display(v00_moaks_shared_kl_drop_mbm_pc_df)
display(Markdown('### 3.1.2 BML Variable Principal Components Loadings'))
display(v00_moaks_shared_kl_drop_mbm_loadings_df.sort_values(by='PC1').head(10))
```

## Results
PCA does not capture shared variance in the data very well in the baseline BML variables. This suggests that using these principal components as predictors offers very little advantage over the original matrix.
